### 데이터프레임 필터 
1. loc[{인덱스의 조건},{컬럼의 조건}]
    - 인덱스의 값을 기준으로 데이터프레임 필터 
    - 컬럼의 값을 기준으로 데이터프레임 필터 
    - 인덱스와 컬럼의 조건식을 이용하여 데이터프레임 필터 
2. iloc[{인덱스의 위치},{컬럼의 위치}]
    - 인덱스나 컬럼의 위치를 기준으로 데이터프레임 필터 
    

```python
# 데이터프레임에서 특정한 데이터를 제거 
# drop({조건}, axis = {0(rows) | 1(columns)}, inplace = {bool})
# 데이터프레임명.drop()
# axis : 기준 -> 행을 지울것이냐? 열을 지울것이냐? 
# inplace : cka / 거짓 형태의 데이터를 입력 
# 참 : 기준이 되는 데이터프레임을 변경 
# 거짓 : 기준이 되는 데이터프레임을 변경하지 않고 결과를 출력 (기본값)

# df에서 Unnamed : 0 컬럼을 제거 -> 컬럼에 이름이 없는 경우 Unnamed로 자동 표기 
#df = df.drop('Unnamed: 0' , axis = 1)
df.drop('Unnamed: 0' , axis = 1 , inplace = True)
```

```python
# 데이터 프레임의 인덱스를 재지정(초기화)
df.reset_index(drop = True , inplace = True)
```

```python
df.head(3)
```

```python
# 인덱스의 조건이 1이고 칼럼이 조건이 총사망자인 데이터와 
# 인덱스의 조건이 0이고 컬럼의 조건이 총사망자인 데이터의 차이 -> 인덱스 1의 일일사망자 

df.loc [1,'총사망자'] - df.loc[0, '총사망자']

# 위의 작업은 인덱스의 개수 - 1만큼 반복 

# 일일 사망자가 추가될 리스트를 생성 
# 첫날의 데이터는 0으로 지정 
_list = [0]

for i in range(len(df)-1) : 
    data = df.loc[i + 1, '총사망자'] - df.loc[i, '총사망자']
    _list.append(data)


_list
```

```python
len(_list)
```

```python
df['일일사망자'] = _list 
```

```python
df.head(3)
```

```python
df['총사망자']
```

```python
# 스리즈 형태에 value 를 인덱스 이용 
# shift(n) : n의 기본값은 1 

df['총사망자'].shift()
```

```python
df['일일사망자2'] = (df['총사망자'] - df['총사망자'].shift(1)).fillna(0)
```

```python
df.head(3)
```

```python
## 스리즈 데이터에서 데이터간의 차이를 출력하는 함수 
# diff(n) : n만큼 인덱스가 이동한 데이터와의 차이를 출력 (n = 1 기본값)
df['일일사망자3'] = df['총사망자'].diff().fillna(0)
```

```python
df.head(5)
```

```python
# 일일 확진자 파생변수 생성 
# 일일확진자 -> shift()을 이용하여 생성 
# 일일확진자2 -> diff()을 이용하여 생성 

df['일일확진자'] = (df['총확진자'] - df['총확진자'].shift()).fillna(0)
df['일일확진자'] = df['총확진자'].diff().fillna(0)
df.head(4)
# describe()함수를 이용하여 데이터의 이상이 있는 확인 
```
