```python
import requests
from bs4 import BeautifulSoup as bs 
```

```python
# 요청해보기 

res = requests.get("http://www.naver.com")
```

```python
print(res.text)
```

```python
html_data = res.text
```

```python
# html 문서에서 네이버를 찾아보자

html_data.find('네이버')
```

```python
html_data[378:410]
```

```python
# BeautifulSoup을 이용하여 태그들을 객체화 
soup = bs(html_data, 'html.parser')
```

- 태그의 이름을 기준으로 검색 
    - soup.태그명 : HTML 문서에서 해당 태그의 첫번재 정보를 출력 
    - soup.태그명.string : 첫번째 태그에서 문자만 출력
    - soup.태그명['속성명'] : 첫번재 태그에서 속성의 값을 출력 

```python
soup.a
```

```python
soup.a.string
```

```python
soup.img
```

```python
soup.a['href']
```

- find({태그명})
    - HTML 문자열에서 해당 태그의 첫번재 정보를 출력
    - find(속성명 = 속성의 값) : 태그들 중에서 해당 속성명과 속성값인 첫번째 태그의 정보를 출력 
- find_all({태그명})
    - HTML 문자열에서 해당 태그의 모든 정보를 리스트의 형태로 출력 
    - attrs 매개변수 : {속성명 : 속성값}을 넣어주면 해당 태그중들 속성명과 속성값이 같은 태그들을 선택 
    - limit 매개변수 : 리스트의 길이의 제한 

```python
a_list = soup.find_all('a')
```

```python
a_data = soup.find('a')
```

```python
print(type(a_data))
print(type(a_list))
```

```python
a_data.find('span').get_text()
```

```python
a_list[1].find('span')
# 태그를 여러개 모아놓은 것이 리절트셋.. 
```

```python
# a_list에 있는 모든 문자를 새로운 배열에 담아보자 

a_result = [] 

for i in a_list : 
    # print(type(i))
     # print(i.get_text())
    a_result.append(i.get_text())
a_result
```
